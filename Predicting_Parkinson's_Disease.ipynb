# **FEATURE SELECTION**

```python
parkDF2 = parkDF.copy()
```

```python
#we need to rearrange our columns to take the label column to the far right
Cols = parkDF2.columns.tolist()
LabelIndex = Cols.index('status')
ax = Cols.pop(LabelIndex)
Cols.append(ax)
#display(Cols)
parkDF2 = parkDF2.reindex(columns = Cols)
display(parkDF2)

```

```python
#Assigning numerical values to binary and categorical columns
#number = LabelEncoder()
#parkDF2['name'] = number.fit_transform(parkDF2['name'])
parkDF2['name'] = LabelEncoder().fit_transform(parkDF2['name'])
parkDF2
```

```python
#we need to create a copy of the dataframe then convert the negative column into positive using the abs()
parkDF3 = parkDF2.copy()
parkDF3['spread1'] = abs(parkDF3['spread1'])
#parkDF3
```

```python
X = parkDF3.iloc[:,0:23] #All features
Y = parkDF3.loc[:,['status']] #label class
#display(Features)

#using the chi square test to perform feature selection
#Bfeatures = SelectKBest(score_func = chi2, k = 10)
#fit = Bfeatures.fit(X,Y)

fitt = SelectKBest(score_func = chi2, k = 10).fit(X,Y)
#fitt.scores_
#X.columns
```

```python
# we create a dataframe for the best features and their respective scores for visualization
TbelScore = pd.DataFrame(fitt.scores_)
TbelFeatures = pd.DataFrame(X.columns)
#display(TbelFeatures)
#display(TbelScore)
#Now we merge the two series into a dataframe
GeneralTabel = pd.concat([TbelFeatures, TbelScore], axis = 1)
GeneralTabel.columns = ['Features', 'Scores']
#display(GeneralTabel)
GeneralTabel.sort_values(by = 'Scores', ascending = False)
```

**From the table above, the most informative features are presented in a top down approach. It can also be said that the features with the most dependencies on the class label are presented top-down**


## **NOW WE BUILD THE MODEL**

## USING SELECTED (TOP) FEATURES

```python

x = parkDF3[['MDVP:Flo(Hz)','MDVP:Fo(Hz)','MDVP:Fhi(Hz)','name', 'HNR', 'spread1']]
y = parkDF3[['status']]
#display(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 6)

model = XGBClassifier()
model.fit(x_train, y_train)
```

## **NOW WE EVALUATE THE MODEL**

```python
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred)*100)
```

```python
#Defining metrics i.e., the Predicted probability of the positive class
y_pred_proba = model.predict_proba(x_test) [::,1]
#print(y_pred_proba)
```

```python
#calculating the true positive and false positive rates
#this is calculated using the actual true test class i.e., x_test and the y_predicted probability
FalsePositiveRate, TruePositiveRate, _ = metrics.roc_curve(y_test, y_pred_proba)

#calculate the AUC to see the model's performance
auc = metrics.roc_auc_score(y_test, y_pred_proba)
display(auc)
```

```python
#creating the Curve
Fig, ax = plt.subplots(figsize = (12,6))
plt.plot(FalsePositiveRate, TruePositiveRate, label = 'AUC='+ str(auc), color = 'red')
plt.title('ROC Curve', fontsize = 14)
plt.xlabel('False Positive Rate', fontsize = 12)
plt.ylabel('True Positive Rate', fontsize = 12)
plt.legend(loc = 4)
```

**The area under the curve (AUC) is 0.84, which is very close to one, meaning that the model did a good job.**
